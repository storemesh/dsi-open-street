```python
from dsi_walking_map import WalkerMap
import matplotlib.pyplot as plt
```

```python
# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏û‡∏¥‡∏Å‡∏±‡∏î ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£
lat, lon = 14.072134236615362, 100.60826137548777

# ‡∏™‡∏£‡πâ‡∏≤‡∏á Object ‡πÄ‡∏û‡∏µ‡∏¢‡∏á‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÄ‡∏î‡∏µ‡∏¢‡∏ß input lat, lon
obj = WalkerMap(lat, lon)
```

```python
# 1. ‡πÅ‡∏™‡∏î‡∏á‡∏†‡∏≤‡∏û‡∏ñ‡∏ô‡∏ô (Street)
print("Plotting Street layer...")
fig_street, ax_street = obj.street()
plt.show()
```

```python
# 2. ‡πÅ‡∏™‡∏î‡∏á‡∏†‡∏≤‡∏û‡∏™‡∏¥‡πà‡∏á‡∏õ‡∏•‡∏π‡∏Å‡∏™‡∏£‡πâ‡∏≤‡∏á (Building)
print("Plotting Building layer...")
fig_building, ax_building = obj.building()
plt.show()
```

```python
# 3. ‡πÅ‡∏™‡∏î‡∏á‡∏†‡∏≤‡∏û‡πÅ‡∏´‡∏•‡πà‡∏á‡∏ô‡πâ‡∏≥ (Water)
print("Plotting Water layer...")
fig_water, ax_water = obj.water()
plt.show()
```

```python
# --- ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ method ‡πÉ‡∏´‡∏°‡πà basemap() ‡∏Å‡∏≤‡∏£‡∏£‡∏ß‡∏°‡∏ó‡∏∏‡∏Å layer : ‡∏ñ‡∏ô‡∏ô, ‡∏ô‡πâ‡∏≥, ‡∏™‡∏¥‡πà‡∏á‡∏Å‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á ---
print("Plotting the basemap without service pins...")
fig_base, ax_base = obj.basemap(area=False)
plt.show()
```

```python
# --- ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ method ‡πÉ‡∏´‡∏°‡πà basemap() ‡∏Å‡∏≤‡∏£‡∏£‡∏ß‡∏°‡∏ó‡∏∏‡∏Å layer : ‡∏ñ‡∏ô‡∏ô, ‡∏ô‡πâ‡∏≥, ‡∏™‡∏¥‡πà‡∏á‡∏Å‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á ‡πÄ‡∏û‡∏¥‡πà‡∏° area 15-minutes city ---
print("Plotting the basemap without service pins...")
fig_base, ax_base = obj.basemap(area=True)
plt.show()
```
